```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#櫻坂のメンバーページの中に''守屋麗奈''の文字列があるか検証

#画像読み込み
img = cv2.imread('sakurazaka.jpg',0)
img2 = img.copy()
template = cv2.imread('template_face.jpg',0)
w, h = template.shape[::-1]

# All the 6 methods for comparison in a list
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for meth in methods:
    img = img2.copy()
    method = eval(meth)

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    #類似度の最大最小の値、場所
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    print(min_val, max_val, min_loc, max_loc)
    # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
    #左上と右下の座標を特定すると矩形がかける
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)

    cv2.rectangle(img,top_left, bottom_right, (0,255,0), 2)

#     plt.subplot(121),plt.imshow(res,cmap = 'gray')
#     plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
#     plt.subplot(122),plt.imshow(img,cmap = 'gray')
#     plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
#     plt.suptitle(meth)

    print(meth)
    plt.imshow(img, cmap='gray')
    plt.title('Detected Point')
    
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.gray()
```

```python
#画像の読み込み、グレースケール化
#ターゲット画像
img=cv2.imread('tpl_match.png')
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#パターン画像
template=cv2.imread('tpl_match_tpl.png')
template_gray=cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
```

```python
#画像の表示
plt.imshow(img_gray)
plt.imshow(template_gray)
```

```python
#ターゲット画像に対して、パターン画像の類似度の確認
#res=cv2.matchTemplate(img_gray, template_gray, cv2.TM_CCOEFF)
res=cv2.matchTemplate(img_gray, template_gray, cv2.TM_CCOEFF_NORMED)
res.shape
```

```python
#類似度の高い部分を検出、閾値0.8以上
threshold=0.9
loc=np.where(res>=threshold)
```

```python
#パターン画像の高さと幅を抽出
h,w=template_gray.shape
h,w
```

```python
#検出した部分に矩形を描く
for pt in zip(*loc[::-1]):
    cv2.rectangle(img, (h,w), (pt[0]+w,pt[1]+h), (0,0,255), 2)
```

```python
#画像の保存
cv2.imwrite('tpl_img_detected.png', img)
```

```python
import cv2

capture = cv2.VideoCapture("./Kyoko.mov") # 動画へのパス
fps = capture.get(cv2.CAP_PROP_FPS)/5
count = 0  # フレーム用カウンタ

while True:
    ret, frame = capture.read()
    if not ret:  # 読み込めなかった場合はループを抜ける
        break
    if int(count % fps) == 0:  # 1秒ごとにフレームを保存
        path_num = int(count // fps)
        cv2.imwrite(f"./images/{path_num:02}.png", frame)
    count += 1
```
